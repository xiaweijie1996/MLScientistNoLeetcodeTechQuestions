# Python Iterators & `__iter__` Summary


## 1. Dunder Methods
Dunder (double underscore) methods are special functions like:
- `__init__`
- `__len__`
- `__iter__`

They are called automatically by Python.

Example:
```python
len(x) → x.__len__()
for x in obj → obj.__iter__()
```



## 2. Iterable vs Iterator

**Iterable**:
- Has `__iter__()`
- Can be looped over

**Iterator**:
- Has `__iter__()` and `__next__()`
- Produces values one by one


````

nums = [1, 2, 3]

it = iter(nums)

print(next(it))
print(next(it))
print(next(it))
````


## 3. What `__iter__` does

`__iter__` tells Python how to START iteration.

```python
for x in obj:
```

becomes:

```python
it = obj.__iter__()
next(it)
```



## 4. Example: returning a NEW iterator (like list)


````

class MyList:
    def __init__(self):
        self.data = [10, 20, 30]

    def __iter__(self):
        return iter(self.data)

obj = MyList()

for x in obj:
    print(x)
````


## 5. Example: returning `self` (iterator object)


````

class Counter:
    def __init__(self):
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        self.i += 1
        if self.i <= 3:
            return self.i
        else:
            raise StopIteration

c = Counter()

for x in c:
    print(x)
````


## 6. Important Behavior

This will NOT run twice:


````

for x in c:
    print(x)  # already exhausted
````


## 7. Key Insight

- `return self` → object IS the iterator → single-use
- `return new iterator` → reusable

## Final takeaway

`__iter__` = "give me something I can loop over"
